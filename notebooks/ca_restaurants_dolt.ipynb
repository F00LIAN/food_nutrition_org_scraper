### Manual Approach to Adding Tags

```python
italian = ['Italian', 'Pizza', 'Italian food', 'Italian restaurant', 'Italian cuisine', 'Italian restaurant', 'olive garden']

american = ['Wendys', 'Sports', 'Burger King', 'McDonalds', 'KFC', 'Taco Bell', 'Pizza Hut', 'Subway', 'Domino\'s Pizza', 'Papa Johns', 'Little Caesars', 'Sonic', 'Dairy Queen', 'Arby\'s', 'Jack in the Box', 'Chick-fil-A', 'Wendys', 'Burger King', 'McDonald\'s', 'KFC', 'Taco Bell', 'Pizza Hut', 'Subway', 'Domino\'s Pizza', 'Papa John\'s', 'Little Caesars', 'Sonic', 'Dairy Queen', 'Arby\'s', 'Jack in the Box', 'Chick-fil-A']

mexican = ['mexican', 'mexican food', 'mariscos', 'chipotle', 'latin', 'tortilla', 'taco']

asian = ['Asia', 'tofu', 'woo', 'ramen', 'pho','Noodle', 'Chong', 'Sushi', 'Thai', 'Vietnamese', 'Japanese', 'Korean', 'Chinese', 'Indian', 'Philippine', 'Malaysian', 'Indonesian', 'Singaporean', 'Thai', 'Vietnamese', 'Japanese', 'Korean', 'Chinese', 'Indian', 'Philippine', 'Malaysian', 'Indonesian', 'Singaporean']

# Create a new column called "cuisine_type" that categorizes restaurants based on keywords in their names
# Function to classify cuisine type based on restaurant name
def classify_cuisine(restaurant_name):
    name_lower = restaurant_name.lower()
    
    # Check each cuisine type
    if any(keyword.lower() in name_lower for keyword in italian):
        return 'Italian'
    elif any(keyword.lower() in name_lower for keyword in american):
        return 'American'
    elif any(keyword.lower() in name_lower for keyword in mexican):
        return 'Mexican'
    elif any(keyword.lower() in name_lower for keyword in asian):
        return 'Asian'
    else:
        return 'Other'

# Apply the classification function
df_irvine['cuisine_type'] = df_irvine['restaurant_name'].apply(classify_cuisine)

# Display the results
print("Cuisine Type Distribution:")
print(df_irvine['cuisine_type'].value_counts())
print("\nSample of classified restaurants:")
print(df_irvine[['restaurant_name', 'cuisine_type']].head(10)) 
```

```python
df_irvine.head()
```

```python
# Create a visualization of cuisine types
plt.figure(figsize=(10, 6))

# Count cuisine types
cuisine_counts = df_irvine['cuisine_type'].value_counts()

# Create bar plot
sns.barplot(x=cuisine_counts.values, y=cuisine_counts.index, palette='Set2')
plt.title('Distribution of Cuisine Types in Irvine Restaurants', fontsize=16, fontweight='bold')
plt.xlabel('Number of Restaurants', fontsize=12)
plt.ylabel('Cuisine Type', fontsize=12)
plt.tight_layout()

# Add value labels on bars
for i, v in enumerate(cuisine_counts.values):
    plt.text(v + 0.1, i, str(v), va='center', fontweight='bold')

plt.show()

# Show some examples of each cuisine type
print("\nExamples by cuisine type:")
for cuisine in df_irvine['cuisine_type'].unique():
    print(f"\n{cuisine} restaurants:")
    examples = df_irvine[df_irvine['cuisine_type'] == cuisine]['restaurant_name'].unique()[:5]
    for example in examples:
        print(f"  - {example}")
```

```python
# Create a composite column called restaurant_name_location that is the restaurant name and the location.
df_irvine['restaurant_name_location'] = df_irvine['restaurant_name'] + ' - ' + df_irvine['identifier']

# Find the number of menu items per restaurant_name_location
# Each row represents a menu item, so we count the rows for each location
menu_items_per_location = df_irvine.groupby('restaurant_name_location').size().reset_index(name='total_menu_items')

# Sort by number of menu items (descending)
menu_items_per_location = menu_items_per_location.sort_values('total_menu_items', ascending=False)

print("Top 20 restaurant locations with most menu items:")
print(menu_items_per_location.head(20))

print(f"\nSummary statistics:")
print(f"Total unique restaurant locations: {len(menu_items_per_location)}")
print(f"Average menu items per location: {menu_items_per_location['total_menu_items'].mean():.1f}")
print(f"Median menu items per location: {menu_items_per_location['total_menu_items'].median():.1f}")
print(f"Max menu items at single location: {menu_items_per_location['total_menu_items'].max()}")
print(f"Min menu items at single location: {menu_items_per_location['total_menu_items'].min()}")
```
